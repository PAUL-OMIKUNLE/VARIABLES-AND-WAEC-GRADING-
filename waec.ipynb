Creating a new variable “average_score” represented from “math”, “reading” and “writing” scores.

```python
data["average_score"] = (data["reading score"]+data["writing score"]+data ["math score"])/3
```

```python
data
```

Creating another variable “average_score_cat” that categorises the “average_score” using WAEC grading system. THIS CAN PREDICT THE AVERAGE_SCORE_CAT of any student.

```python
grades = {
    85:"A1", 
    70:"B2", 
    65:"B3", 
    60:"C4", 
    55:"C5",
    50:"C6", 
    45:"D7", 
    40:"E8", 
    0:"F9",
}
def grade(score) :
    for key, letter in grades.items():
        if score>=key:
            return letter
waec_grade= data ["average_score"].map(grade)
data["average_score_cat"] = pd.Categorical (waec_grade, categories=grades.values(), ordered=True)
```

```python
data
```

Relationship between other variables and average_score_cat

```python
sns.relplot(x= 'math score', y='writing score', hue= 'average_score_cat', data=data)
```
